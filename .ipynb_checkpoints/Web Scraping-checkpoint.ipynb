```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver  
import pandas as pd
import time 
```

```python
url = 'https://welcomesaudi.com/restaurant'
page = requests.get(url)
print(page.status_code)
```

```python
driver = webdriver.Chrome()
url = 'https://welcomesaudi.com/restaurant'
columns_ = ['Name','Cuisines', 'City', 'Rating', 'NumberOfReviews']
df = pd.DataFrame(columns=columns_)

for i in range(2, 58):

    # request the page 
    driver.get(url)
    time.sleep(5)
    # extract the body and parse it
    soup = BeautifulSoup(driver.page_source, "html.parser")

    # extrace prices and ars names from page
    
    Name = soup.find_all('h5', {'class':"item-title"})
    Cuisines = soup.find_all('div', {'class':'cuisines'})
    City = soup.find_all('div', {'class':'location'})
    Rating = soup.find_all('span', {'class':'rate text-success'})
    NumberOfReviews = soup.find_all('span', {'class':'review'})
    
    
    clean_names = []
    clean_cuisines = []
    clean_City = []
    clean_ratings = []
    clean_NumberOfReviews = []
    
    for n, c, s, r, o in zip( Name, City, Cuisines, Rating, NumberOfReviews):
        clean_names.append(n.text.strip())
        clean_City.append(c.text)
        clean_cuisines.append(s.text)
        clean_ratings.append(r.text)
        clean_NumberOfReviews.append(o.text)
        
       

        
    # store them 
    new_rows_df = pd.DataFrame(
        {
                'Name': clean_names,
                'City' : clean_City,
                'Cuisines' : clean_cuisines,
                'Rating' : clean_ratings,
                'NumberOfReviews': clean_NumberOfReviews
                
                

        })

    df = pd.concat([df, new_rows_df], ignore_index=True)

    # prepare new link
    url = f'https://welcomesaudi.com/restaurant?query=&page={i}'
driver.quit()
```
